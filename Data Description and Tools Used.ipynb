## Data to be used in the project:

For the Toronto neighborhood data, a Wikipedia page exists that has all the information we need to explore and cluster the neighborhoods in Toronto. ## Libraries to be used to develop the Project:

#### Pandas:

Library written for the Python programming language for data manipulation and analysis

#### Folium:

Folium builds on the data wrangling strengths of the Python ecosystem and the mapping strengths of the leaflet.js library. It helps us in manipulatinh the data in Python and then visualize it in on a Leaflet map via folium acting as a Library.

#### Scikit Learn: 

Built on NumPy, SciPy, and matplotlib, Scikit Learn helps in Clustering of unlabelled data through the module sklearn.cluster. 

#### JSON: 

JavaScript Object Notation (JSON) is a standard text-based format for representing structured data based on JavaScript object syntax. Here, JSON library of python is used to handle JSON format files.

#### XML: 

XML stands for extensible Markup Language. It is a markup language much like HTML and it has been designed to store and transport data in plain text format.

#### Geocoder: 

It is an Apache2 Licensed Geocoding library, written in Python. It is used on the data set for fetching the location data.

#### Beautiful Soap 4: 

Beautiful Soup 4 is a Python library for pulling data out of HTML and XML files. When worked in conjunction with Parser, it provides idiomatic ways of navigating, searching, and modifying the parse tree.

#### Matplotlib: 

It is a comprehensive library for creating static, animated, and interactive visualizations in Python by acting as a plotting library for the Python programming language and its numerical mathematics extension NumPy.

## Foursquare API:

The Foursquare API allows application developers to interact with the foursquare platform. The API itself is a RESTful set of addresses to which you can send requests, so there's really nothing to download onto your server. There are both GET and POST methods you can use, which means you aren't just limited to reading from feeds, you can also do useful stuff like checking in and creating locations using the API. 

The Foursquare API is a RESTful API based on HTTP requests and JSON responses. This version of the API, version 2, uses OAuth 2.0. This means that all requests will need to be encrypted and sent via HTTPS. To explore the cities, we will use the Venue Recommendation which returns a list of recommended venues near a certain location. In order to make the query in the Foursquare API, we need the coordinates given in Latitude and Longitude for a given Neighbourhood.
