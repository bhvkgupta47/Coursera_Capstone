# The Battle of Neighbourhoods | Business Proposal

### Introduction:

The purpose of this project is to explore the neighbourhoods of North York, Canada. Neighbourhood empowerment is salient to citizens and that’s why, the idea of empowering people to shape or make decisions in their neighbourhoods is gathering momentum around the world. If the research is done well, then there are multiple benefits for corporates as well as local citizens:

•	Improve the quality of decision-making and levels of citizen engagement

•	Helping to rebuild legitimacy and make bureaucracies more responsive

•	Develop new sources of social value. 

If the research is done badly, then it can lead to conflict, growing inequalities and failing public services.

You know what they say about buying or renting a home: location, location, location. Neighbourhood matters a lot when it comes to how satisfied people are in their house or apartment and it can be a real deal breaker (or deal-maker) when people are searching for the neighbourhoods. And while it’s hard to really know a place until you live there, taking the time to research a neighbourhood before you move can help ensure that you don’t accidentally end up with a big case of buyer’s or renter’s remorse later on. Digging into the details of a neighbourhood before committing to living there can save you a lot of headaches later on.

### Location:

North York is an eclectic, multicultural district home to the hands-on Ontario Science Centre and the Aga Khan Museum, with exhibits on Islamic culture in a striking modern building. In the area’s north, Black Creek Pioneer Village is an 1800s living museum. Sprawling Downsview Park includes a lake, event spaces, and a flea and farmers’ market, while Edwards Gardens has a greenhouse, fountains, and botanic gardens.
